```python
import pandas as pd
```

```python
url = "https://raw.githubusercontent.com/MiguelCortezPino/etl-data-pipeline-1701472022/refs/heads/main/data/raw/E_inventario.csv"
df = pd.read_csv(url)

df.head()
```

```python
#Exploracion de datos
df.shape
df.columns
df.info()
df.isnull().sum()
```

```python
#limpieza de datos

print("Filas con valores nulos en 'id_inventario':")
display(df[df['id_inventario'].isnull()])

print("\nFilas con valores nulos en 'fecha':")
display(df[df['fecha'].isnull()])

print("\nFilas con valores nulos en 'cantidad_movimiento':")
display(df[df['cantidad_movimiento'].isnull()])

# Limpiar 'cantidad_movimiento': eliminar 'uds' y convertir a numérico
df['cantidad_movimiento'] = df['cantidad_movimiento'].astype(str).str.replace(' uds', '', regex=False)
df['cantidad_movimiento'] = pd.to_numeric(df['cantidad_movimiento'], errors='coerce')

# Convertir 'fecha' a formato de fecha
df['fecha'] = pd.to_datetime(df['fecha'], errors='coerce')

df.info()
```

```python
#motivo de rechazo
#df['motivo_rechazo'].value_counts() # Esta línea causó el error porque la columna no existe.

# Para ver las columnas disponibles en el DataFrame, puedes usar:
print(df.columns)
```

```python
#creacion de archivos csv reject y curated

# Definir las columnas críticas para la validación
critical_columns = ['id_inventario', 'fecha', 'cantidad_movimiento']

# Identificar filas con valores nulos en alguna de las columnas críticas
df_rechazados = df[df[critical_columns].isnull().any(axis=1)].copy()

# Identificar filas sin valores nulos en ninguna de las columnas críticas
df_validos = df[~df[critical_columns].isnull().any(axis=1)].copy()

```

```python
# Guardar los DataFrames en archivos CSV
df_validos.to_csv("E_inventario_cureted.csv", index=False)
df_rechazados.to_csv("E_inventario_rejects.csv", index=False)

print("Archivos guardados con éxito: 'E_inventario_cureted.csv' y 'E_inventario_rejects.csv'")
```
